```python
import json



choice=1
while (choice!=0):
    print('===============================================================================')
    choice=int(input('''Press 1.Search word
Press 2.Update the word
Press 3.Add new word
Press 4.Display all words
Press 0.Exit
'''))
    if(choice==4):
        with open('dictionary.txt','r')as f:
            a=json.load(f)
            b=sorted(a)
            for i in b:
                print(i,' : ',a[i])
    elif(choice==1):
        word=input("Enter word to search: ").title()
        with open('dictionary.txt','r')as f:
            a=json.load(f)
            for i in a:
                if(word==i):
                    print(i)
                    print('Meaning of :',word,'is :',a[i])
                    break
            else:
                print('No word in dictionary you can try adding')
    elif(choice==3):
        temp={}
        temp2={}
        word=input("Enter the new word: ").title()
        with open('dictionary.txt','r')as f:
            a=json.load(f)
            for i in a:
                if(word==i):
                    print("Word already in the dictionary")
                    break
            else:
                meaning=input("Enter the meaning: ").title()
                temp[word]=meaning
                for i in a,temp:
                    temp2.update(i)
                with open('dictionary.txt','w')as g:
                    json.dump(temp2,g)
    elif(choice==0):
        print("Thank you")
    elif(choice==2):
        temp={}
        temp2={}
        word=input("Enter the word to update: ").title()
        with open('dictionary.txt','r')as f:
            a=json.load(f)
            for i in a:
                
                if(word==i):
                    meaning=input("Enter meaning").title()
                    temp[word]=meaning
                    break
            else:
                print(word,'is not in a list try adding')
        for i in a,temp:
            temp2.update(i)
        with open('dictionary.txt','w')as f:
            json.dump(temp2,f)
    else:
        print('Please choose between given option')
        
                    
        
```
